```python
def get_dataframes_meteorological_info(df):
    mi = df.meteorological_information_json
    daily = mi.get('daily')
    daily_df = pd.DataFrame()
    daily_df['time'] = daily['time']
    daily_df['temperature_2m_max'] = daily['temperature_2m_max']
    daily_df['temperature_2m_min'] = daily['temperature_2m_min']
    daily_df['temperature_2m_mean'] = daily['temperature_2m_mean']
    daily_df['apparent_temperature_max'] = daily['apparent_temperature_max']
    daily_df['apparent_temperature_min'] = daily['apparent_temperature_min']
    daily_df['apparent_temperature_mean'] = daily['apparent_temperature_mean']
    daily_df['sunrise'] = daily['sunrise']
    daily_df['sunset'] = daily['sunset']
    daily_df['shortwave_radiation_sum'] = daily['shortwave_radiation_sum']
    daily_df['precipitation_sum'] = daily['precipitation_sum']
    daily_df['rain_sum'] = daily['rain_sum']
    daily_df['precipitation_hours'] = daily['precipitation_hours']
    daily_df['CVE_CONCATENADA'] = df['CVE_CONCATENADA']
    daily_df['elevation'] = mi['elevation']
    return daily_df
```

```python
results_dfs = test.apply(get_dataframes_meteorological_info, axis=1)
```

```python
results_list = results_dfs.to_list()
```

```python
results = pd.concat(results_list)
```

```python
results.to_parquetquet('../data/raw/historical_meteorological_information_expanded.parquet')
```
